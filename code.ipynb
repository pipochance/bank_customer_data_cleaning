```python
#Task 3: Monthly Repayment for each customer 
repay_df.columns
```

```python
# Step 1: Fix the typo in 'Costomer'
repay_df.rename(columns={'Costomer': 'Customer'}, inplace=True)

# Step 2: Convert 'Month' column to datetime
repay_df['Month'] = pd.to_datetime(repay_df['Month'])

# Step 3: Create a new 'Year_Month' column for grouping
repay_df['Year_Month'] = repay_df['Month'].dt.to_period('M')

# Step 4: Group by 'Customer' and 'Year_Month', then sum 'Amount'
monthly_repayment = repay_df.groupby(['Customer', 'Year_Month'])['Amount'].sum().reset_index()

# Step 5: Rename 'Amount' to 'Monthly_Repayment' for clarity
monthly_repayment.rename(columns={'Amount': 'Monthly_Repayment'}, inplace=True)

# Step 6: Preview
monthly_repayment.head()
```

```python
#Task 4: Top 10 Highest Paying Customers

#Group by 'Customer' and sum total repayment
total_repayment = repay_df.groupby('Customer')['Amount'].sum().reset_index()

#sort in descending order
top_10_customers = total_repayment.sort_values(by='Amount', ascending=False).head(10)

#Rename column for clarity
top_10_customers.rename(columns={'Amount': 'Total_Repayment'}, inplace=True)

#preview
top_10_customers
```

```python
#Lets visualize that thang
import matplotlib.pyplot as plt
import seaborn as sns

# Set the plot style
sns.set(style="whitegrid")

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Total_Repayment',
    y='Customer',
    data=top_10_customers,
    palette='viridis'
)

# Add labels and title
plt.xlabel('Total Repayment Amount')
plt.ylabel('Customer ID')
plt.title('Top 10 Highest Paying Customers')
plt.tight_layout()
plt.show()
```

```python
#Task 5: Which segment is spending more money?

# Step 1: Fix column names for consistency
cust_df.rename(columns={'Customer': 'Customer_ID'}, inplace=True)
spend_df.rename(columns={'Costomer': 'Customer'}, inplace=True)

# Step 2: Merge customer info with spend data
spend_merged = pd.merge(spend_df, cust_df, left_on='Customer', right_on='Customer_ID', how='left')

# Step 3: Group by 'Segment' and sum the 'Amount'
segment_spend = spend_merged.groupby('Segment')['Amount'].sum().reset_index()

# Step 4: Sort descending to find top segments
segment_spend = segment_spend.sort_values(by='Amount', ascending=False)

# Step 5: Preview
segment_spend
```

```python
#Visual
plt.figure(figsize=(8, 5))
sns.barplot(data=segment_spend, x='Amount', y='Segment', palette='magma')
plt.title('Total Spend by Segment')
plt.xlabel('Total Spend')
plt.ylabel('Customer Segment')
plt.tight_layout()
plt.show()
```

```python
# Task 6: Which age group is spending more money?

# Merge
spend_with_age = pd.merge(spend_df, cust_df, left_on='Customer', right_on='Customer_ID', how='left')

# Bin ages
age_bins = [17, 25, 35, 45, 55, 65, 100]
age_labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '65+']
spend_with_age['Age_Group'] = pd.cut(spend_with_age['Age'], bins=age_bins, labels=age_labels)

# Group and fix observed warning
age_group_spend = spend_with_age.groupby('Age_Group', observed=True)['Amount'].sum().reset_index()

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=age_group_spend, x='Amount', y='Age_Group', palette='coolwarm', legend=False)
plt.title('Total Spend by Age Group')
plt.xlabel('Total Spend')
plt.ylabel('Age Group')
plt.tight_layout()
plt.show()
```

```python
#Task 7: Which Is the Most Profitable Segment?

# Merge spend and repayment with customer data
spend_with_segment = pd.merge(spend_df, cust_df, left_on='Customer', right_on='Customer_ID', how='left')
repay_with_segment = pd.merge(repay_df, cust_df, left_on='Customer', right_on='Customer_ID', how='left')

# Group and sum spend and repayment by segment
segment_spend = spend_with_segment.groupby('Segment')['Amount'].sum().reset_index()
segment_spend.rename(columns={'Amount': 'Total_Spend'}, inplace=True)

segment_repay = repay_with_segment.groupby('Segment')['Amount'].sum().reset_index()
segment_repay.rename(columns={'Amount': 'Total_Repayment'}, inplace=True)

# Merge spend and repayment data
segment_profit = pd.merge(segment_spend, segment_repay, on='Segment')
segment_profit['Profit'] = segment_profit['Total_Spend'] - segment_profit['Total_Repayment']

# Sort by profit
segment_profit = segment_profit.sort_values(by='Profit', ascending=False)

# Display result
segment_profit

```

```python
#Task8: In which category are customers spending more money?

# Group by 'Type' (category) and sum the 'Amount'
category_spend = spend_df.groupby('Type')['Amount'].sum().reset_index()

# Sort descending by amount
category_spend = category_spend.sort_values(by='Amount', ascending=False)

# Display result
category_spend

plt.figure(figsize=(10,6))
sns.barplot(data=category_spend, x='Amount', y='Type', palette='viridis')
plt.title('Total Spending by Category')
plt.xlabel('Total Amount Spent')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
#Task 9: Impose an interest rate of 2.9% for each customer on any due amount

# Step 1: Ensure 'Year_Month' in monthly_spend is Period type (monthly)
monthly_spend['Year_Month'] = monthly_spend['Year_Month'].apply(
    lambda x: x.to_period('M') if isinstance(x, pd.Timestamp) else x
)

# Step 2: Group repay_df by Customer and Month, sum the Amount
monthly_repay = repay_df.groupby(['Customer', 'Month'])['Amount'].sum().reset_index()

# Convert 'Month' in monthly_repay to Period type (monthly) and rename to Year_Month
monthly_repay['Month'] = monthly_repay['Month'].apply(
    lambda x: x.to_period('M') if isinstance(x, pd.Timestamp) else x
)
monthly_repay.rename(columns={'Month': 'Year_Month', 'Amount': 'Monthly_Repay'}, inplace=True)

# Step 3: Merge monthly spend and repayment data on Customer and Year_Month
monthly_data = pd.merge(monthly_spend, monthly_repay, on=['Customer', 'Year_Month'], how='outer')

# Step 4: Fill NaN values with 0 (for months with no spend or repay)
monthly_data['Monthly_Spend'] = monthly_data['Monthly_Spend'].fillna(0)
monthly_data['Monthly_Repay'] = monthly_data['Monthly_Repay'].fillna(0)

# Step 5: Calculate due amount (Spend - Repay)
monthly_data['Due_Amount'] = monthly_data['Monthly_Spend'] - monthly_data['Monthly_Repay']

# Step 6: Apply interest rate of 2.9% on positive due amounts only
monthly_data['Interest'] = monthly_data['Due_Amount'].apply(lambda x: x * 0.029 if x > 0 else 0)

# Display the first few rows to verify
print(monthly_data.head())
```

```python
#Task 10:Monthly profit for the bank
monthly_profit = monthly_data.groupby('Year_Month')['Interest'].sum().reset_index(name='Monthly_Profit')
monthly_profit.head()
```

```python
import pandas as pd

# Load Excel data
file_path = "Credit_Banking_Project.xlsx"

# Load sheets
spend_df = pd.read_excel(file_path, sheet_name="Spend")
repay_df = pd.read_excel(file_path, sheet_name="Repayment")


data = pd.read_excel('Credit_Banking_Project.xlsx', sheet_name=None)

cust_df=data['Customer Acqusition']
spend_df = data['Spend']
repay_df = data['Repayment']

#renaming for typo correction
spend_df.rename(columns={'Costomer': 'Customer'}, inplace=True)
repay_df.rename(columns={'Costomer': 'Customer'}, inplace=True)

# Convert 'Month' columns to datetime
spend_df['Month'] = pd.to_datetime(spend_df['Month'])
repay_df['Month'] = pd.to_datetime(repay_df['Month'])

# Extract Year-Month
spend_df['Month'] = spend_df['Month'].dt.to_period('M')
repay_df['Month'] = repay_df['Month'].dt.to_period('M')

# Task 6: Monthly spend
monthly_spend = spend_df.groupby(['Customer', 'Month'])['Amount'].sum().reset_index()
monthly_spend.rename(columns={'Month': 'Year_Month', 'Amount': 'Monthly_Spend'}, inplace=True)

# Task 9: Monthly repayment
monthly_repay = repay_df.groupby(['Customer', 'Month'])['Amount'].sum().reset_index()
monthly_repay.rename(columns={'Month': 'Year_Month', 'Amount': 'Monthly_Repay'}, inplace=True)

# Merge spend and repay
monthly_data = pd.merge(monthly_spend, monthly_repay, on=['Customer', 'Year_Month'], how='outer')

# Fill missing values
monthly_data['Monthly_Spend'] = monthly_data['Monthly_Spend'].fillna(0)
monthly_data['Monthly_Repay'] = monthly_data['Monthly_Repay'].fillna(0)

# Task 10: Due amount
monthly_data['Due_Amount'] = monthly_data['Monthly_Spend'] - monthly_data['Monthly_Repay']

# Task 11: Interest
monthly_data['Interest'] = monthly_data['Due_Amount'].apply(lambda x: x * 0.029 if x > 0 else 0)

# Task 12: Bank profit = interest earned
monthly_data['Bank_Profit'] = monthly_data['Interest']

# Optional: convert Period to string for Excel compatibility
monthly_data['Year_Month'] = monthly_data['Year_Month'].astype(str)

# Save to new sheet using openpyxl
with pd.ExcelWriter(file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    monthly_data.to_excel(writer, sheet_name='Monthly_Report', index=False)

print("Excel updated with new sheet: 'Monthly_Report'")
```
